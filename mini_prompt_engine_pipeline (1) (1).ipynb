#  Mini Prompt Engine using LangChain

## Pipeline Flow
User Input → Validation → Prompt Template → Dynamic Prompt → Output

```
# Install (run once)
# !pip install langchain
```

```
from langchain.prompts import PromptTemplate, ChatPromptTemplate
```

  Step 1: Input Validation Layer

```
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience! Defaulting to 'beginner'")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone! Defaulting to 'formal'")
        tone = "formal"

    return audience, tone
```

 Step 2: Prompt Templates

```
teaching_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} teaching style"
)

interview_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Ask interview questions about {topic} for {audience} in a {tone} tone"
)

story_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} storytelling style"
)
```

 Step 3: Prompt Generator Function

```
def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    if style == "teaching":
        return teaching_template.format(topic=topic, audience=audience, tone=tone)
    elif style == "interview":
        return interview_template.format(topic=topic, audience=audience, tone=tone)
    elif style == "storytelling":
        return story_template.format(topic=topic, audience=audience, tone=tone)
    else:
        return "Invalid style selected"
```

 Step 4: Chat Prompt System

```
def chat_prompt(role, topic):
    if role == "teacher":
        system = "You are a helpful teacher."
    elif role == "interviewer":
        system = "You are a job interviewer."
    elif role == "motivator":
        system = "You are a motivational speaker."
    else:
        system = "You are an assistant."

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system),
        ("user", "Talk about {topic}")
    ])

    return chat_template.format_messages(topic=topic)
```

 Step 5: Run Full Pipeline

```
# User Input
topic = "Neural Networks"
audience = "beginner"
tone = "fun"
style = "storytelling"

# Pipeline Execution
print("=== PIPELINE EXECUTION ===")

validated_audience, validated_tone = validate_inputs(audience, tone)
print("Validated:", validated_audience, validated_tone)

prompt = generate_prompt(topic, validated_audience, validated_tone, style)
print("Generated Prompt:", prompt)

messages = chat_prompt("teacher", topic)
print("\nChat Messages:")
for msg in messages:
    print(msg.type.upper(), ":", msg.content)
```

 Step 6: Reusability Test

```
reusable_template = PromptTemplate(
    input_variables=["topic", "audience"],
    template="Explain {topic} for {audience}"
)

inputs = [
    ("AI", "beginners"),
    ("Python", "kids"),
    ("Machine Learning", "engineers"),
    ("Deep Learning", "researchers"),
    ("Data Science", "students")
]

print("=== REUSABILITY TEST ===")
for t, a in inputs:
    print(reusable_template.format(topic=t, audience=a))
```
